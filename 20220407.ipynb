```python
import numpy as np
import matplotlib.pyplot as plt

x1= np.linspace(0,14,50)
y1= np.sin(x1)
y2= np.sin(x1+2)

fig = plt.figure(figsize=(10,6))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(x1,y1)
ax2.plot(x1,y2)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

#pivot (인덱스, 컬럼, 값)
fgb = flights.pivot('month','year', 'passengers')

plt.figure(figsize=(10,8))

sns.heatmap(fgb, annot=True,fmt = 'd') #cmp= 을 통해 색깔을 정할 수 있음.
plt.show()
```

```python
# stripplot 모든 데이터를 점으로 시각화
sns.stripplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
# stripplot 모든 데이터를 점으로 시각화 
# jitter=True 설정하면 위치를 무작위로 바꿔서 겹치치 않게 만듬
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
plt.show()
```

```python
# folium 지도를 시각화

# 데이터 변수가 여러개인 다차원 데이터는 데이터 종류에 따라 경우 다름

# 1) 분석하고자하는 데이터가 모두 실수인 경우
# 2) 분석하고자하는 데이터가 모두 카테고리 값인 경우
# 3) 분석하고자하는 데이터가 모두 실수값과 카테고리 값이 섞여있는 경우

# jointplot : scatter plot을 사용할 때 seaborn lib.에서 사용하는 것이 jointplot
#             모두실수형인 데이터 일 때 사용 

# jointplot(x= 'xname', y='yname', data= dataframe, kind='종류 ex) scatter')

iris = sns.load_dataset('iris')
sns.jointplot(x= 'sepal_length', y='sepal_width', data= iris, kind='scatter')

plt.show()
iris
```
