```python
 # 한글 font를 쓰기위해 위에 font manager를 import 함

fig = plt.figure(constrained_layout=True)

font_location='C:\Windows\Fonts\HMFMPYUN.TTF' #폰트 위치 저장
font_name= fm.FontProperties(fname=font_location).get_name() # font manager를 이용한 font_name 변수 설정
print(font_name)
matplotlib.rc('font',family=font_name) # font 정보 불러오기

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
x= range(0,10)
y= range(5,15)
a= range(1,11)
b= [1,4,9,16,25,36,49,64,81,100]
ax1.plot(x,y,'ro')
ax2.plot(a,b,'b^')



ax1.set_xlabel('시간')
ax1.set_ylabel('수량')

ax2.set_xlabel('시간')
ax2.set_ylabel('수량')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 막대그래프

plt.style.use('ggplot') #그래프 스타일 default의 그래프 타입을 변경

fig= plt.figure()

ax=fig.add_subplot(111)
x=[3,4]
y=[1,2]

# x축 항목 및 색 지정
labels=['english','math']
colors=['r','b']
ax.bar(x,y,tick_label=labels,color=colors) #bat  = 막대 그래프

plt.show()
```

```python
import matplotlib.pyplot as plt

# 막대그래프

plt.style.use('ggplot') #그래프 스타일 default의 그래프 타입을 변경

fig= plt.figure()

ax=fig.add_subplot(111)
x=[1,2,3]
y=[2,1,4]

# x축 항목 및 색 지정
labels=['english','math','science']
colors=['r','b','y']
ax.bar(x,y,tick_label=labels,color=colors) #bat  = 막대 그래프

for i,j in enumerate(y):
    ax.text(i+1,j+0.05,str(j),color='red',fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt

# 막대그래프

plt.style.use('ggplot') #그래프 스타일 default의 그래프 타입을 변경

fig= plt.figure()

ax=fig.add_subplot(111)
x=[3,4]
y=[1,2]

# x축 항목 및 색 지정
labels=['english','math']
colors=['r','b']
ax.barh(x,y,tick_label=labels,color=colors) #bat  = 막대 그래프

plt.show()
```

```python
import numpy as np

x= [1,2]
y1= [1,2]
y2= [2,5]
y3= [3,7]

fig = plt.figure()

ax= fig.add_subplot(111)
w=0.2
ax.bar(x,y1,width=w,label='a')
ax.bar(np.array(x)+w,y2,width=w,label='b')
ax.bar(np.array(x)+2*w,y3,width=w,label='c')
ax.legend()
plt.show()
```

```python
#pie 차트 만들기
plt.style.use('ggplot')

fig= plt.figure()
ax=fig.add_subplot(111)
label=['english','math','korean']
sizes=[45,35,20]
ax.pie(sizes, labels=label)

plt.show()
```

```python
#pie 차트 만들기
plt.style.use('ggplot')

fig= plt.figure()
ax=fig.add_subplot(111)
label=['english','math','korean']
sizes=[45,35,20]
ax.pie(sizes, labels=label, autopct='%1.1f%%',shadow=True)

plt.show()
```

```python
#pie 차트 만들기
plt.style.use('ggplot')

fig= plt.figure()
ax=fig.add_subplot(111)
label=['english','math','korean']
sizes=[45,35,20]
ax.pie(sizes, labels=label, autopct='%1.1f%%',shadow=True)

#그래프 제목(title) 설정
ax.set_title("subjects")

plt.show()
```

```python
# box plot 그리기

plt.style.use('ggplot')
fig=plt.figure()

ax=fig.add_subplot(111)

x=[1,2,3,3,11,20]
ax.boxplot(x)

plt.show()
```

```python
# 여러개의 box plot 그리기

plt.style.use('ggplot')
fig=plt.figure()

ax=fig.add_subplot(111)

x=[[1,2,3,3,11,20],[3,8,5,4,19,5,0]]
label=['english','math']
ax.boxplot(x,labels=label)

plt.show()
```
